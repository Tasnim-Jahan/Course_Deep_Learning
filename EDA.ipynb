Exploratory Data Analysis (EDA) in Python

```python
import matplotlib.pyplot as plt
```

```python
# Create a pie chart presenting the male/female proportion.
plt.figure(figsize = (6,8))
plt.pie(titanic_df["Sex"].value_counts(),
	autopct="%0.2f%%", labels=["Male", "Female"],
	explode=(0, 0.1))
plt.legend()
plt.show()
```

```python
titanic_df["Survived"].value_counts()
print("Survived = ", (titanic_df["Survived"]==1).sum())
print("Not Survived = ", (titanic_df["Survived"]==0).sum())
```

```python
plt.figure(figsize = (6,8))
plt.pie(titanic_df["Survived"].value_counts(),
	autopct="%0.2f%%", labels=["Not Survived", "Survived"],
	colors=["red", "green"], explode=(0, 0.1))
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (6,8))
plt.pie(titanic_df["Embarked"].value_counts(), autopct="%0.2f%%",
	labels=["Southampton","Cherbourg","Queenstown"],
	colors=['red','green','blue'], explode=(0.,0.1,0.1))
plt.legend()
plt.show()
```

KMeans Clustering with Scikit-learn in Python

```python
from google.colab import files
uploaded = files.upload()
```

```python
mall_customer_df = pd.read_csv(io.BytesIO(uploaded["Mall_Customers.csv"]))
mall_customer_df
```

```python
mall_customer_df.loc[:,["Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
X = mall_customer_df.loc[:,["Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter=100, n_init = 10)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
y_clusters = kmeans.fit_predict(X)
y_clusters.astype
```

```python
import numpy as np
x_array = np.array(X)
```

```python
# Visualization
plt.figure(figsize=(15,8))
sns.scatterplot( x = x_array[ y_clusters == 0, 0], y = x_array[y_clusters == 0, 1], label = 'Cluster1', s = 60)
sns.scatterplot( x = x_array[ y_clusters == 1, 0], y = x_array[y_clusters == 1, 1], label = 'Cluster2', s = 60)
sns.scatterplot( x = x_array[ y_clusters == 2, 0], y = x_array[y_clusters == 2, 1], label = 'Cluster3', s = 60)
```

```python
sns.scatterplot( kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker = 'x', label = 'Centroids', color = 'black', s = 60)
plt.show()
```
